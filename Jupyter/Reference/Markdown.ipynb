## Char formatting

To make text **bold** surround it with two asterisks `**`. 

To make text *italic* use a single asterisk `*` or underscore `_`.

`_**Bold** inside italics_ and **vice-_versa_** also work.` 

_**Bold** inside italics_ and **vice-_versa_** also work. 

`~~Strikethrough~~ uses two tildes`

~~Strikethrough~~ uses two tildes

`monospace` (such as code) uses \`backtick\` characters as delimiters.

---

## Indentation

Blocks are indented with \>, and multiple levels of indentation are indicated by repetition: \>\>\> indents three levels.

>First level
>>Second level
>>>And so on

To separate two paragraphs, leave a intermediate blank line

Markdown joins any number of spaces in only one space. To add more spaces use `&nbsp;` or laxex code like `$\qquad$`

---

## Lists

Unordered lists are \* or - at the beginning of a line. Lists can be nested by indenting using two spaces for each level of nesting.

* First element of unordered list
* Second element of unordered list
>* Indented element of unordered list (using \>\*)

Ordered lists are created by typing any number followed by a period at the beginning of a line. 

1. First element of ordered list
2. Second element of ordered list
>2.1 Indented element or ordered list

## $\LaTeX$

You can integrate formulas (MathJax) inside text using $\LaTeX$

Inline formulas are surrounded by `$` characters

$\qquad$ `$y=x^2$` &nbsp; &nbsp; &nbsp; $y=x^2$

You can put stand-alone formulas centered in their own lines using double `$$` characters like in `$$y=x^2$$`

$$y=x^2$$

---

## $\LaTeX$ Basics

$\qquad$ Superscript &nbsp; &nbsp; `$A^B$` &nbsp; &nbsp; $A^B$

$\qquad$ Subcript &nbsp; &nbsp; `$A_B$` &nbsp; &nbsp; $A_B$

$\qquad$ Vector &nbsp; &nbsp; `$\vec{A}$` &nbsp; &nbsp; $\vec{A}$

$\qquad$ Fraction &nbsp; &nbsp; `$\frac{A}{B}$` &nbsp; &nbsp; $\frac{A}{B}$

$\qquad$ Binomial &nbsp; &nbsp; `$\binom{A}{B}$` &nbsp; &nbsp; $\binom{A}{B}$

$\qquad$ Root &nbsp; &nbsp; &nbsp; `$\sqrt[n]A$` &nbsp; &nbsp; $\sqrt[n]A$

$\qquad$ Sum &nbsp;&nbsp; &nbsp; `$\sum_A^B$` &nbsp; &nbsp; $\sum_A^B$

$\qquad$ Product &nbsp;&nbsp; &nbsp; `$\prod_A^B$` &nbsp; &nbsp; $\prod_A^B$

$\qquad$ Integral &nbsp;&nbsp; &nbsp; `$\int_A^B$` &nbsp; &nbsp; $\int_A^B$

$\qquad$ Block {&nbsp;}&nbsp; &nbsp; `$A^{B+C}$` &nbsp; &nbsp; $A^{B+C}$

$\qquad$ Auto sized parenthesis &nbsp;&nbsp; `$\left( \frac{\int_a^bxdx}{2} \right)$` &nbsp;&nbsp; $\bigl( \frac{\int_a^bxdx}{2} \bigr)$

$\qquad$ Big parenthesis &nbsp;&nbsp; `$\bigl( \frac{\int_a^bxdx}{2} \bigr)$` &nbsp;&nbsp; $\left( \frac{\int_a^bxdx}{2} \right)$

It is important to note that the format changes for **inline** formulas, between `$` and for **stand-alone** formulas, between `$$`.

Inline example for summation: `$\sum_{i=0}^n i$` &nbsp; &nbsp; $\sum_{i=0}^n i$

Stand-alone example for summation: `$$\sum_{i=0}^n i$$`

$$\sum_{i=0}^n i$$

If you want to use the **stand-alone** format aligned to the left you can use the **aligned** property:

`$\begin{aligned} \sum_{i=1}^n i \end{aligned}$`

$\begin{aligned} \sum_{i=1}^n i \end{aligned}$

You can insert text inside a $\LaTeX$ construct:

`$$x^2+y^1=r^2\qquad \text{Circle Equation}$$`

$$x^2+y^1=r^2\qquad \text{Circle Equation}$$

## $\LaTeX$ Horizontal braces

We can put braces over or below the objects:

Brace over the object: `\overbrace{a=b}` $\quad \overbrace{a=b}$

Brace under the object: `\underbrace{a=b}` $\quad \underbrace{a=b}$

Complex brace construct:

`$$\overbrace{a+\underbrace{b+c}_\text{inside} +d}^\text{outside}$$`

$$\overbrace{a+\underbrace{b+c}_\text{inside} +d}^\text{outside}$$
